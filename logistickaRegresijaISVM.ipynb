```python
import numpy as np
import pandas as pd
from sklearn.utils import shuffle
from matplotlib import pyplot as plt
import matplotlib.cm as cm
import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
from sklearn import linear_model
from sklearn import model_selection
from sklearn import metrics
from sklearn import preprocessing
from sklearn import datasets
from sklearn import svm
```

```python
flights = pd.read_csv('flights_small.csv')
# flights
```

```python
working_flights = flights[['MONTH', 'DAY', 'DAY_OF_WEEK', 'SCHEDULED_DEPARTURE', 'DEPARTURE_DELAY', 'TAXI_OUT', 'DISTANCE', 'SCHEDULED_ARRIVAL', 'AIRLINE_AA', 'AIRLINE_AS', 'AIRLINE_B6', 'AIRLINE_DL', 'AIRLINE_EV', 'AIRLINE_F9', 'AIRLINE_HA', 'AIRLINE_MQ', 'AIRLINE_NK', 'AIRLINE_OO', 'AIRLINE_UA', 'AIRLINE_US', 'AIRLINE_VX', 'AIRLINE_WN']]
working_flights
```

```python
x = working_flights
y = flights['DELAY']
```

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y, test_size = 0.33, random_state = 7, stratify = y )
```

```python
#skaliranje
scaler = preprocessing.StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
#logisticka regresija
model = linear_model.LogisticRegression()
model.fit(x_train, y_train)
```

```python
y_test_predicted = model.predict(x_test)
y_train_predicted = model.predict(x_train)
train_score = metrics.accuracy_score(y_train, y_train_predicted)
lr_test_acc_score = metrics.accuracy_score(y_test, y_test_predicted)
lr_test_rec_score = metrics.recall_score(y_test, y_test_predicted)
lr_test_f1_score = metrics.f1_score(y_test,y_test_predicted)
print('Train score: ', train_score)
print('Test score: ', lr_test_acc_score)
print('Acc: ',lr_test_acc_score)
print(' rec: ', lr_test_rec_score)
print(' f1: ', lr_test_f1_score)
# ----------------------
```
