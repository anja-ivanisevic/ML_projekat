```python
import numpy as np
import pandas as pd
from sklearn.utils import shuffle
from matplotlib import pyplot as plt
import matplotlib.cm as cm
import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
from sklearn import linear_model
from sklearn import model_selection
from sklearn import metrics
from sklearn import preprocessing
from sklearn import datasets
from sklearn import svm
```

```python
p1 = pd.read_csv('flights.csv', nrows=200000)
```

```python
delay = p1['ARRIVAL_DELAY'] > 15
delay_flights = p1[delay]
delay_flights['DELAY'] = True

not_delay = p1['ARRIVAL_DELAY'] <= 15
not_delay_flights = p1[not_delay]
not_delay_flights['DELAY'] = False

not_delay_flights = not_delay_flights.sample(len(delay_flights))

flights = delay_flights.append(not_delay_flights, ignore_index=True)
flights

flights.to_csv('flights_small.csv')
```

```python
flights = pd.read_csv('flights_small.csv')
flights
```

```python
working_flights = flights[ ['MONTH', 'DAY', 'DAY_OF_WEEK', 'SCHEDULED_DEPARTURE', 'DEPARTURE_DELAY', 'TAXI_OUT', 'DISTANCE', 'SCHEDULED_ARRIVAL']]
# working_flights
```

```python
x = working_flights
y = flights['DELAY']
```

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y, test_size = 0.33, random_state = 7, stratify = y )
```

```python
# logisticka regresija
scaler = preprocessing.StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
model = linear_model.LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_test_predicted = model.predict(x_test)
test_score = metrics.accuracy_score(y_test, y_test_predicted)
print(test_score)
y_train_predicted = model.predict(x_train)
train_score = metrics.accuracy_score(y_train, y_train_predicted)
print(train_score)
# ----------------------
```

```python
sumt = sum(1 for i in y_test if i == True)
print(sumt)
sumf = sum(1 for i in y_test if i == False)
print(sumf)
```
