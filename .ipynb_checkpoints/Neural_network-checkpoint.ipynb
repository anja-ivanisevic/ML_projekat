```python
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras import losses, optimizers
from keras.utils.vis_utils import plot_model
```

```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
from sklearn import preprocessing
from sklearn import datasets
from sklearn import model_selection
from sklearn import metrics
```

```python
# Ucitavanje podataka
p = pd.read_csv('flights_small.csv', low_memory=False)
```

```python
p
```

```python
p.columns
```

```python
x = p[['MONTH', 'DAY', 'DAY_OF_WEEK', 'SCHEDULED_DEPARTURE', 'DEPARTURE_DELAY', 'TAXI_OUT', 'DISTANCE', 'SCHEDULED_ARRIVAL', 'AIRLINE_AA', 'AIRLINE_AS', 'AIRLINE_B6', 'AIRLINE_DL', 'AIRLINE_EV', 'AIRLINE_F9', 'AIRLINE_HA', 'AIRLINE_MQ', 'AIRLINE_NK', 'AIRLINE_OO', 'AIRLINE_UA', 'AIRLINE_US', 'AIRLINE_VX', 'AIRLINE_WN']]
```

```python
x
```

```python
y = p[['DELAY']]
```

```python
# Podela podataka na trening i test skup
x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y, test_size = 0.33, random_state = 7)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
number_of_features = x_train.shape[1]
output_size = 1
```

```python
# Pretprocesiranje
scaler = preprocessing.StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
# Neuronska mreza za klasifikaciju
```

```python
model = Sequential()
```

```python
model.add(Dense(units= 50, input_dim = number_of_features, activation = 'relu'))
```

```python
model.add(Dense(units=40, activation='relu'))
```

```python
model.add(Dense(units = output_size, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss=losses.binary_crossentropy, metrics=['accuracy'])
```

```python
model.summary()
```

```python
# Obucavanje modela
history = model.fit(x_train, y_train, epochs= 30, batch_size=16, verbose=1, validation_split=0.3)
```

```python
epochs = history.epoch
```

```python
loss = history.history['loss']
validation_loss = history.history['val_loss']
```

```python
plt.xlabel('epochs')
plt.ylabel('loss')
plt.plot(epochs, loss, c='red', label='training')
plt.plot(epochs, validation_loss, c='orange', label='validation')
plt.legend(loc='best')
plt.show()
```

```python
acc = history.history['acc']
validation_acc = history.history['val_acc']
```

```python
plt.xlabel('epochs')
plt.ylabel('acc')
plt.plot(epochs, acc, c='red', label='training')
plt.plot(epochs, validation_acc, c='orange', label='validation')
plt.legend(loc='best')
plt.show()
```

```python
# Evaluacija mreze
test_loss, test_acc = model.evaluate(x_test, y_test)
print ("Test loss: {}, test accuracy: {}".format(test_loss, test_acc))
```

```python
# Doslo je do preprilagodjavanja, vidimo da su nam rezultati za validacioni skup znatno losiji od trening skupa
# Broj epoha - 5
```

```python
model = Sequential()
```

```python
model.add(Dense(units= 50, input_dim = number_of_features, activation = 'relu'))
```

```python
model.add(Dense(units=40, activation='relu'))
```

```python
model.add(Dense(units = output_size, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss=losses.binary_crossentropy, metrics=['accuracy'])
```

```python
history = model.fit(x_train, y_train, epochs= 5, batch_size=16, verbose=1, validation_split=0.3)
```

```python
epochs = history.epoch
loss = history.history['loss']
validation_loss = history.history['val_loss']
```

```python
plt.xlabel('epochs')
plt.ylabel('loss')
plt.plot(epochs, loss, c='red', label='training')
plt.plot(epochs, validation_loss, c='orange', label='validation')
plt.legend(loc='best')
plt.show()
```

```python
acc = history.history['acc']
validation_acc = history.history['val_acc']
```

```python
plt.xlabel('epochs')
plt.ylabel('acc')
plt.plot(epochs, acc, c='red', label='training')
plt.plot(epochs, validation_acc, c='orange', label='validation')
plt.legend(loc='best')
plt.show()
```

```python
y_test_predicted = model.predict(x_test)
```

```python
y_test_predicted = (y_test_predicted > 0.5)
y_test_predicted
```

```python
# Evaluacija mreze
test_loss, test_acc = model.evaluate(x_test, y_test)
print ("Test loss: {}, test accuracy: {}".format(test_loss, test_acc))
```

```python
# Odziv
recall = metrics.recall_score(y_test, y_test_predicted)
recall
```

```python
# Preciznost
precision = metrics.precision_score(y_test, y_test_predicted)
precision
```

```python
# F1 mera
f1 = metrics.f1_score(y_test, y_test_predicted)
f1
```
